# Watermark

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['image.cmap'] = 'gray'
```

```python
logo=cv2.imread("logo_black.jpg")
img =cv2.imread('image.jpg')
plt.imshow(logo[:,:,::-1])
plt.show()
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
rows,cols,ch=logo.shape
img_h, img_w, _ = img.shape

# Cx1 and Cy1 are centre of the left half
cx = int(img_w/2)
cy = int(img_h/2)

# tlc : top left corner
tlc_x = int(cx - cols/2)
tlc_y = int(cy - rows/2)

# brc : bottom right corner
brc_x = int(cx + cols/2)
brc_y = int(cy + rows/2)

roi = img[tlc_y:brc_y, tlc_x:brc_x]
```

```python
logo_gray=cv2.cvtColor(logo,cv2.COLOR_BGR2GRAY)
ret,mask=cv2.threshold(logo_gray,10,255,cv2.THRESH_BINARY)
mask_inv=cv2.bitwise_not(mask)
plt.imshow(mask_inv)

```

```python
bg = cv2.bitwise_and(roi,roi,mask = mask_inv)
fg = cv2.bitwise_and(logo,logo,mask=mask)
```

```python
plt.imshow(bg[:,:,::-1])
```

```python
plt.imshow(bg[:,:,::-1])
```

```python
plt.imshow(fg[:,:,::-1])
```

```python
dst=cv2.add(fg,bg)
img[tlc_y:brc_y, tlc_x:brc_x]=dst
# cv2.resize(wm,(300,100))
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
logo_white = cv2.imread('logo_white.jpg')
img2 = cv2.imread('image2.jpg')
plt.imshow(logo_white[:,:,::-1])
plt.show()
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
rows,cols,ch=logo_white.shape
img_h, img_w, _ = img2.shape

# Cx1 and Cy1 are centre of the left half
cx = int(img_w/2)
cy = int(img_h/2)

# tlc : top left corner
tlc_x = int(cx - cols/2)
tlc_y = int(cy - rows/2)

# brc : bottom right corner
brc_x = int(cx + cols/2)
brc_y = int(cy + rows/2)

roi2 = img2[tlc_y:brc_y, tlc_x:brc_x]
```

```python
logo_gray=cv2.cvtColor(logo_white,cv2.COLOR_BGR2GRAY)
ret,mask=cv2.threshold(logo_gray,127,255,cv2.THRESH_BINARY_INV)
mask_inv=cv2.bitwise_not(mask)
plt.imshow(mask)
plt.show()
plt.imshow(mask_inv)
plt.show()
```

```python
bg = cv2.bitwise_and(roi2,roi2,mask = mask_inv)
fg = cv2.bitwise_and(logo_white,logo_white,mask=mask)
```

```python
plt.imshow(bg[:,:,::-1])
```

```python
plt.imshow(fg[:,:,::-1])
```

```python
dst2=cv2.add(fg,bg)
img2[tlc_y:brc_y, tlc_x:brc_x]=dst2
# cv2.resize(wm,(300,100))
plt.imshow(img2[:,:,::-1])
plt.show()
```
